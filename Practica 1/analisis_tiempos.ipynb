```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
c_safe_server = pd.read_csv("./csock/safe-server.csv")
c_safe_client = pd.read_csv("./csock/safe-client.csv")
java_time = pd.read_csv("./javasock/time.csv")
```

```python
c_time = c_safe_server.append(c_safe_client)
```

```python
means = pd.DataFrame([["java_all", java_time["time"].mean()/1000000]], columns=["name","mean"])
means = means.append(pd.DataFrame([["java_read",java_time[java_time["operation"]=="read"]["time"].mean()/1000000]], columns=["name","mean"]))
means = means.append(pd.DataFrame([["java_write",java_time[java_time["operation"]=="write"]["time"].mean()/1000000]], columns=["name","mean"]))
means = means.append(pd.DataFrame([["c_all",c_time["time"].mean()]], columns=["name","mean"]))
means = means.append(pd.DataFrame([["c_read",c_time[c_time["operation"]=="read"]["time"].mean()]], columns=["name","mean"]))
means = means.append(pd.DataFrame([["c_write",c_time[c_time["operation"]=="write"]["time"].mean()]], columns=["name","mean"]))
means.plot.bar(x='name', y='mean', rot=0)
```
