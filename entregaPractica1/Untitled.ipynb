```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ccsv = pd.read_csv("csock/time.csv")
jcsv = pd.read_csv("javasock/time.csv")
```

```python
jcsv['time'] = jcsv['time'].apply(lambda x: x/1_000_000_000)
```

```python
ccsv['key'] = "c";
jcsv['key'] = "java";
```

```python
DF = pd.concat([ccsv,jcsv], sort=False)
DF
```

```python
DF.groupby(['key']).mean().unstack().plot.bar()
```

```python
DF.groupby(['key']).std().unstack().plot.bar()
```
